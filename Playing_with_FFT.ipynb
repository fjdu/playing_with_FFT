```python
def gaussian(A, x0, sigma):
    def f(x):
        return A * np.exp(-0.5*((x-x0)/sigma)**2)
    return f

def gaussian2d(A, x0, y0, sigma_x, sigma_y):
    def f(x, y):
        return A * np.exp(-0.5*(((x-x0)/sigma_x)**2 +
                                ((y-y0)/sigma_y)**2))
    return f

def spiral(nspiral, r2theta, I_r, w_dr):
    def f(xx, yy, zz, xcen, ycen):
        X = xx - xcen
        Y = yy - ycen
        r = np.sqrt(X*X + Y*Y)
        theta = np.arctan2(Y, X)
        dtheta = 2.0 * np.pi / nspiral
        I_r_s = I_r(r)
        theta_s = r2theta(r)
        for i in xrange(nspiral):
            theta_this = theta_s + i*dtheta
            dx_this = r * np.cos(theta_this) - X
            dy_this = r * np.sin(theta_this) - Y
            dr_this = np.sqrt(dx_this**2 + dy_this**2)
            zz = zz + I_r_s * w_dr(dr_this)
        return zz
    return f

def fft_round_trip(y):
    # The returned vector should be almost identical to y.
    return np.fft.ifft(np.fft.fft(y))

def fft_dewave(ffted, freq, t0):
    return ffted * np.exp((2.0 * np.pi * 1j * t0) * freq)

def easy_fft_1d(x, y):
    dx = x[1] - x[0]
    n = len(x)
    freq = fft.fftfreq(n, d=dx)
    y_fft = fft.fft(y) * np.exp(2.0j*np.pi*freq*(x[0]-dx))
    return (fft.fftshift(freq),
            fft.fftshift(y_fft))

def easy_fft_2d(x, y, z):
    dx, dy = x[1] - x[0], y[1] - y[0]
    nx, ny = len(x), len(y)
    freqx, freqy = (fft.fftfreq(nx, d=dx),
                    fft.fftfreq(ny, d=dy))
    fxx, fyy = np.meshgrid(freqx, freqy)
    z_fft = fft.fft2(z) * np.exp(2.0j*np.pi*(fxx*(x[0]-dx) +
                                             fyy*(y[0]-dy)))
    return (fft.fftshift(freqx),
            fft.fftshift(freqy),
            fft.fftshift(z_fft))

def test_easy_fft_1d(num = 101):
    x = linspace(-50,50, num=num)
    y = gaussian(A=0.5, x0=0, sigma=2.0)(x)

    f, yf = easy_fft_1d(x, y)

    figure(figsize=(15,4))

    plt.subplot(131,title='image')
    plot(x, y)
    plt.subplot(132, title='fft-real')
    plot(f, np.real(yf))
    plt.subplot(133, title='fft-imag')
    plot(f, np.imag(yf))
    return

def test_easy_fft_2d(nx = 101, ny=202):
    x = linspace(-50, 50, num=nx)
    y = linspace(-30, 30, num=ny)
    xx, yy = np.meshgrid(x, y)
    zz = gaussian2d(A=0.5, x0=0, y0=0, sigma_x=1.0, sigma_y=3.0)(xx, yy)
    fx, fy, fz = easy_fft_2d(x, y, zz)

    figure(figsize=(6,10))

    plt.subplot(131, aspect='equal', title='image')
    contourf(x, y, zz)
    plt.subplot(132, aspect='equal', title='fft-real')
    contourf(fx, fy, np.real(fz))
    plt.subplot(133, aspect='equal', title='fft-imag')
    contourf(fx, fy, np.imag(fz))
    print np.max(np.abs(np.real(fz)))
    print np.max(np.abs(np.imag(fz)))
    return

def test_easy_fft_2d_mix(nx = 401, ny=502):
    x = linspace(-50, 50, num=nx)
    y = linspace(-50, 50, num=ny)
    xx, yy = np.meshgrid(x, y)
    zz = gaussian2d(A=7.5, x0=-12, y0=7, sigma_x=1.5, sigma_y=3.0)(xx, yy) + \
         gaussian2d(A=20.0, x0=9, y0=5, sigma_x=2.0, sigma_y=6.6)(xx, yy) + \
         gaussian2d(A=9.0, x0=4, y0=-12, sigma_x=8.0, sigma_y=2.0)(xx, yy) + \
         np.random.randn(*xx.shape)
    #zz = np.sin(xx*1.0 + yy*3.0) + gaussian2d(A=0.5, x0=0, y0=0, sigma_x=1.0, sigma_y=3.0)(xx, yy)
    fx, fy, fz = easy_fft_2d(x, y, zz)

    figure(figsize=(15,15))
    
    nlev = 100

    plt.subplot(131, aspect='equal', title='image')
    contourf(x, y, zz, nlev)
    plt.subplot(132, aspect='equal', title='fft-real')
    contourf(fx, fy, np.real(fz), nlev)
    plt.subplot(133, aspect='equal', title='fft-imag')
    contourf(fx, fy, np.imag(fz), nlev)
    print np.max(np.real(fz)), np.min(np.real(fz))
    print np.max(np.imag(fz)), np.min(np.imag(fz))
    return

def test_easy_fft_2d_spiral(nx = 401, ny=401, nspiral=2):
    x = linspace(-50, 50, num=nx)
    y = linspace(-50, 50, num=ny)
    xx, yy = np.meshgrid(x, y)
    zz = np.zeros_like(xx) #np.random.randn(*xx.shape)
    
    def r2theta(r):
        theta = (0.6*np.pi) * np.log(r*0.01+1e-4)
        return theta
    def I_r(r):
        return np.exp(-r*0.06)
    def w_dr(dr):
        return np.exp(-(dr*0.2)**2)
    
    zz = spiral(nspiral, r2theta, I_r, w_dr)(xx, yy, zz, 0.0, 0.0)
    
    fx, fy, fz = easy_fft_2d(x, y, zz)

    figure(figsize=(15,15))
    
    nlev = 100

    plt.subplot(131, aspect='equal', title='image')
    contourf(x, y, zz, nlev)
    plt.subplot(132, aspect='equal', title='fft-real')
    contourf(fx, fy, np.real(fz), nlev)
    ax = gca()
    ax.set_xlim((-0.5,0.5))
    ax.set_ylim((-0.5,0.5))
    plt.subplot(133, aspect='equal', title='fft-imag')
    contourf(fx, fy, np.imag(fz), nlev)
    print np.max(np.real(fz)), np.min(np.real(fz))
    print np.max(np.imag(fz)), np.min(np.imag(fz))
    return
```

```python
def polar2complex(mag, angle):
    return mag * np.exp(1j*angle)

def randomize_phase(y, disturb_scale=0.0):
    y_fft = np.fft.fft(y)
    mag = np.abs(y_fft)
    ang = np.angle(y_fft) + disturb_scale * np.random.uniform(low=0.0,
                                                              high=2.0*np.pi, size=y.shape)
    y_fft_modi = polar2complex(mag, ang)
    return np.fft.ifft(y_fft_modi)

def get_signal_from_spectrum(mag, angle):
    return fft.fftshift(fft.ifft(fft.fftshift(polar2complex(mag, angle))))

def test1():
    num = 150
    x = linspace(-50,50, num=num)
    y = gaussian(A=0.5, x0=0, sigma=10.0)(x) + 0.1 * np.sin(x/5)
    figure(figsize=(15,7))
    plot(x, y, lw=3, label='0')
    for ds in [0.2, 0.5, 2.0, 10.0]:
        ynew = np.abs(randomize_phase(y, disturb_scale=ds))
        print np.max(abs(y-ynew))
        plot(x, ynew, label='{:.1f}'.format(ds))
    legend()
    return

def test2():
    num = 1024
    x = linspace(-50,50, num=num)
    sp = gaussian(A=0.5, x0=3.0, sigma=0.001)(x) - gaussian(A=0.5, x0=-3.0, sigma=0.001)(x)
    y = get_signal_from_spectrum(sp, linspace(0, 2*np.pi, num=len(sp)))
    figure(figsize=(15,7))
    plot(np.real(y))
    plot(np.imag(y))
    return
```

```python
test2()
```
